```python
import pandas as pd
import numpy as np

pd.options.display.max_rows = 1000
```

```python
df = pd.read_csv("/Users/irsyad/Desktop/Portfolio/credit-risk-lending-club/data/processed/accepted_cleaned.csv", low_memory = False)

df.shape
df.head(1).T
```

```python
df.dtypes
```

```python
date_cols = ['issue_d', 'earliest_cr_line', 'last_pymnt_d', 'last_credit_pull_d']

for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
default_statuses = [
    'Charged Off',
    'Late (31-120 days)',
    'Late (16-30 days)',
    'In Grace Period',
    'Default'
]

df['is_default'] = df['loan_status'].isin(default_statuses).astype(int)

df['is_default'].value_counts()
```

```python
df['vintage'] = df['issue_d'].dt.to_period('M')

df['vintage'].value_counts().sort_index().head(20)
```

```python
df.to_csv('/Users/irsyad/Desktop/Portfolio/credit-risk-lending-club/data/processed/accepted_features.csv', index=False)
print("Saved successfully")
```
